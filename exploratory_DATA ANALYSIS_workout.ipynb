# 2- Grouping


```python
df.head()
```

```python
df['HARDSHIP_INDEX'] = df['HARDSHIP_INDEX'] + 10
```

```python
bins2 = np.linspace(min(df['PERCENT_AGED_16__UNEMPLOYED']) , max(df['PERCENT_AGED_16__UNEMPLOYED']), 4)
bins2
```

```python
groups2 = ( 'young' , 'middle' , 'old')
```

```python
df['age_pinned']= pd.cut(df['PERCENT_AGED_16__UNEMPLOYED'], bins2 , labels = groups2 , include_lowest = True )
```

```python
df.head()
```

```python
df['age_pinned']
```

```python
df_test = df[['income_pinned', 'age_pinned' , 'HARDSHIP_INDEX']]
```

```python
df_group = df_test.groupby(['income_pinned', 'age_pinned'], as_index = False).mean()
```

```python
df_group
```

```python
df_pivot = df_group.pivot(index = 'income_pinned' , columns = 'age_pinned')
```

```python
df_pivot
```

### visualizing pivot table as heatmap

```python
fig, ax = plt.subplots()
im = ax.pcolor(df_pivot, cmap='RdBu')

#label names
row_labels = df_pivot.columns.levels[1]
col_labels = df_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(df_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(df_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

# 4 - ANOVA analysis of variance

```python
sns.barplot(x = 'price' , y = 'make', data = df2)
```

```python
df.columns
```

```python
df_anova = df[['HARDSHIP_INDEX' , 'income_pinned']]
df_anova_gp = df_anova.groupby(df['income_pinned'])
```

```python
df_anova_gp.head()
```

```python
df_anova_gp.get_group("medium")["HARDSHIP_INDEX"]
```

```python
f_val, p_val = stats.f_oneway(df_anova_gp.get_group("low")['HARDSHIP_INDEX'], df_anova_gp.get_group('medium')['HARDSHIP_INDEX'], df_anova_gp.get_group('high')['HARDSHIP_INDEX'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```

# note that ANOVA works on categorical values
